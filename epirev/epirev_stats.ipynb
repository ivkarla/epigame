```python
import pandas as pd

# Define the file path
file_path = "/home/kivi/Downloads/"
file_name = "epirev_extracted_data_210324.xlsx"

# Load the Excel file into a Pandas DataFrame
excel_data = pd.read_excel(file_path+file_name, sheet_name=None)

# Display the sheet names to ensure successful loading
print("Sheet names in the Excel file:", list(excel_data.keys()))
```

To convert the dtype of a column from object to int, we need to ensure that all values in the column can be converted to integers. If there are any non-numeric values or missing values (NaN), the conversion will fail.

```python
design_details_df = excel_data["Design Details"]
column = "Cohort Size - Total\n[Question ID: 57634][Field ID: 93697x122082]"

if design_details_df[column].dtype == 'object':
    # Convert the column to numeric, coercing errors to NaN
    design_details_df[column] = pd.to_numeric(design_details_df[column], errors='coerce')
    # Convert NaN values to 0
    design_details_df[column] = design_details_df[column].fillna(0)
    # Convert the column to int
    design_details_df[column] = design_details_df[column].astype(int)

# Now, the column should be converted to integer dtype
```

Plot cohort size distribution.

```python
import matplotlib.pyplot as plt

# Select the "Design Details" sheet from the loaded Excel data
design_details_df = excel_data["Design Details"]
column = "Cohort Size - Total\n[Question ID: 57634][Field ID: 93697x122082]"
plt.figure(figsize=(8, 4))
design_details_df[column].hist(bins=50, color='skyblue', edgecolor='black')
plt.title("Distribution of Cohort Size")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)

# Calculate the median, Q1, and Q3
median_value = design_details_df[column].median()
Q1 = design_details_df[column].quantile(0.25)
Q3 = design_details_df[column].quantile(0.75)

# Plot the vertical lines for median, Q1, and Q3
plt.axvline(median_value, color='red', linestyle='-', linewidth=2, label=f'Median: {median_value:.2f}')
plt.axvline(Q1, color='red', linestyle='--', linewidth=2, label=f'Q1: {Q1:.2f}')
plt.axvline(Q3, color='red', linestyle='--', linewidth=2, label=f'Q3: {Q3:.2f}')

plt.legend()
plt.show()
```

The code below first filters out rows that contain 0, NA, or None values in both columns and calculates the percentage of studies that meet this condition. Then, it filters rows where there is an integer in both columns and calculates the percentage of studies where only good outcome patients are considered. Finally, it calculates the good to poor outcome ratio for these studies and plots its distribution.

```python
# Filter rows with both 0 or NaN values in both columns (non-operated cohorts)
filtered_df = design_details_df[(design_details_df["Cohort Size - Good outcomes\n[Question ID: 57634][Field ID: 93761x122146]"].isin([0.0])) &
                                (design_details_df["Cohort Size - Poor outcomes\n[Question ID: 57634][Field ID: 93762x122147]"].isin([0.0])) |
                                (design_details_df["Cohort Size - Good outcomes\n[Question ID: 57634][Field ID: 93761x122146]"].isna()) &
                                (design_details_df["Cohort Size - Poor outcomes\n[Question ID: 57634][Field ID: 93762x122147]"].isna())]
# Calculate the percentage of studies
percentage_non_operated = (len(filtered_df) / len(design_details_df)) * 100
print(f"Percentage of studies using non-operated patients undergoing iEEG monitoring: {percentage_non_operated:.2f} %")

# Filter only rows with numeric values
filtered_df_integers = design_details_df[(pd.to_numeric(design_details_df["Cohort Size - Good outcomes\n[Question ID: 57634][Field ID: 93761x122146]"], errors='coerce').notnull()) &
                                         (pd.to_numeric(design_details_df["Cohort Size - Poor outcomes\n[Question ID: 57634][Field ID: 93762x122147]"], errors='coerce').notnull())]

# Calculate the percentage of studies with good outcome patients only
filtered_df_good_outcome = filtered_df_integers[(filtered_df_integers["Cohort Size - Good outcomes\n[Question ID: 57634][Field ID: 93761x122146]"] > 0.0) &
                                                (filtered_df_integers["Cohort Size - Poor outcomes\n[Question ID: 57634][Field ID: 93762x122147]"] == 0.0)]
percentage_good_outcome_only = (len(filtered_df_good_outcome) / len(design_details_df)) * 100
print(f"Percentage of studies considering only good outcome patients: {percentage_good_outcome_only:.2f} %")

# Calculate the good to poor outcome ratio for the studies meeting the condition
both_outcomes = filtered_df_integers.drop(filtered_df_good_outcome.index)

outcome_ratios = both_outcomes["Cohort Size - Good outcomes\n[Question ID: 57634][Field ID: 93761x122146]"] / both_outcomes["Cohort Size - Poor outcomes\n[Question ID: 57634][Field ID: 93762x122147]"]

# Plot the distribution of good to poor outcome ratio
plt.figure(figsize=(4, 3))
outcome_ratios.hist(bins=50, color='skyblue', edgecolor='black')
plt.title("Distribution of Good to Poor Outcome Ratio")
plt.xlabel("Good to Poor Outcome Ratio")
plt.ylabel("Frequency")
plt.grid(True)

# Calculate the median, Q1, and Q3
median_value = outcome_ratios.median()
Q1 = outcome_ratios.quantile(0.25)
Q3 = outcome_ratios.quantile(0.75)

# Plot the vertical lines for median, Q1, and Q3
plt.axvline(median_value, color='red', linestyle='-', linewidth=2, label=f'Median: {median_value:.2f}')
plt.axvline(Q1, color='red', linestyle='--', linewidth=2, label=f'Q1: {Q1:.2f}')
plt.axvline(Q3, color='red', linestyle='--', linewidth=2, label=f'Q3: {Q3:.2f}')

plt.legend()
plt.show()
```

The code below reads the possible options directly from the "Cohort Selection Criteria" column and then print a table and plot a pie chart to visualize these percentages. Adjust the DataFrame df to match your actual dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Initialize an empty dictionary to store counts of each option
option_counts = {}

# Iterate through each row and update the counts
for row in design_details_df["Cohort Selection Criteria\n[Question ID: 57635][Field ID: 97180x126675]"]:
    if isinstance(row, str):  # Handle single strings
        # Split the string by ", " and check for specific options
        row_options = [option.strip() for option in row.split("\n")]
        
        # Handle cases where "Consecutive" is present
        if any(option == "Consecutive" for option in row_options):
            option_counts["Consecutive"] = option_counts.get("Consecutive", 0) + 1
        # Handle cases where "Unclear" is present along with other options
        elif any(option == "Unclear" for option in row_options):
            option_counts["Unclear"] = option_counts.get("Unclear", 0) + 1
        # Handle cases where "Non-consecutive" and "Non-operated" are present together
        elif all(option in row_options for option in ["Non-consecutive", "Non-operated"]):
            option_counts["Non-consecutive"] = option_counts.get("Non-consecutive", 0) + 1
        else:
            # Handle all other cases
            for option in row_options:
                option_counts[option] = option_counts.get(option, 0) + 1

# Convert counts to percentages
total_rows = len(design_details_df)
option_percentages = {option: (count / total_rows) * 100 for option, count in option_counts.items()}

# Create a DataFrame for visualization
percentage_design_details_df = pd.DataFrame(option_percentages.items(), columns=["Option", "Percentage"])

# Print the table
print("Percentage of each option in Cohort Selection Criteria:")
print(percentage_design_details_df)

# Plot the pie chart
plt.figure(figsize=(4, 4))
plt.pie(percentage_design_details_df["Percentage"], labels=percentage_design_details_df["Option"], autopct='%1.1f%%', startangle=140)
plt.title("Cohort Selection Criteria")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

##### Let us move of to the "Risk of Bias Assessment" sheet. This sheet contains questions of the QUAPAS questionaire for the RoB assessment as columns. 
##### Each row contains an answer, which are for some questions Yes, No or Unclear; and for other questions Risk: Low, Risk: High and Unclear. We revise the Excel columns M, O, Q, S, U, X, Y, Z, AB, AD, AF, from AH to AM, from AO to AT, and from AV to AY. 
##### We make a pie chart of answer percentages across studies for each question.

```python
import openpyxl

# Sheet
rob_df = excel_data["Risk of Bias Assessment"]

# Specify the columns to be revised
columns_to_revise = ["M", "O", "Q", "S", "U", "X", "Y", "Z", "AB", "AD", "AF", "AH", "AI", "AJ", "AK", "AL", "AM", "AO", "AP", "AQ", "AR", "AS", "AT", "AV", "AW", "AX", "AY"]

def column_letter_to_index(column_letter):
    """
    Convert Excel column letter to index (0-based).
    """
    return openpyxl.utils.column_index_from_string(column_letter) - 1


# Iterate over each column and create pie charts
for column in columns_to_revise:
    # RoB question / column title
    q = rob_df.columns[column_letter_to_index(column)]

    # Calculate percentages of each answer
    answer_counts = rob_df[q].value_counts(normalize=True) * 100

    # Plot pie chart
    plt.figure(figsize=(4, 4))
    plt.pie(answer_counts, labels=answer_counts.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'{q.split("[")[0]}')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
    plt.show()
```

Below, we visualize a Venn diagram of the studies based on the properties: cohort selection (consecutive or not), surgery performed (operated/non-operated/mixed), patient outcome (only seizure-free/both), and ground truth (RA/SOZ/Alternative/combinations).

```python
from matplotlib_venn import venn3

# Extract relevant columns for study characteristics
consecutiveness = design_details_df["Cohort Selection Criteria\n[Question ID: 57635][Field ID: 97180x126675]"].str.contains("Consecutive", na=False)
good_outcome_only = design_details_df[(design_details_df["Cohort Size - Good outcomes\n[Question ID: 57634][Field ID: 93761x122146]"] > 0.0) &
                                    (design_details_df["Cohort Size - Poor outcomes\n[Question ID: 57634][Field ID: 93762x122147]"] == 0.0)]
nonop_only = design_details_df[(design_details_df["Cohort Size - Good outcomes\n[Question ID: 57634][Field ID: 93761x122146]"].isin([0.0])) &
                                (design_details_df["Cohort Size - Poor outcomes\n[Question ID: 57634][Field ID: 93762x122147]"].isin([0.0])) |
                                (design_details_df["Cohort Size - Good outcomes\n[Question ID: 57634][Field ID: 93761x122146]"].isna()) &
                                (design_details_df["Cohort Size - Poor outcomes\n[Question ID: 57634][Field ID: 93762x122147]"].isna())]
ground_truth = design_details_df["Ground Truth\n[Question ID: 57640][Field ID: 93704x122089]"]

# Categorize studies based on characteristics
consecutive_set = set(design_details_df[consecutiveness].index)
good_outcome_only_set = set(good_outcome_only.index)
both_outcomes_set = set(design_details_df.copy().drop(good_outcome_only.index).index)
nonop_set = set(nonop_only.index)
op_set = set(design_details_df.copy().drop(nonop_only.index).index)
ground_truth_res_set = set(list(design_details_df[ground_truth == "Resection Area"].index) + list(design_details_df[ground_truth == "Seizure Onset Zone\nResection Area"].index) + list(design_details_df[ground_truth == "Resection Area\nSeizure Onset Zone"].index))
ground_truth_soz_set = set(design_details_df[ground_truth == "Seizure Onset Zone"].index)

# Creating the Venn Diagram
plt.figure(figsize=(10, 8))
venn = venn3(
    [consecutive_set, op_set, ground_truth_res_set],
    set_labels=("Consecutive Cohort", "Operated Patients", "Ground Truth: Resection Area")
)
venn.get_label_by_id('100').set_text(len(consecutive_set))
venn.get_label_by_id('010').set_text(len(op_set))
venn.get_label_by_id('110').set_text(len(consecutive_set.intersection(op_set)))
plt.title("Venn Diagram: Study Design")
plt.show()
```
