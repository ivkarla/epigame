The code below first filters out rows that contain 0, NA, or None values in both columns and calculates the percentage of studies that meet this condition. Then, it filters rows where there is an integer in both columns and calculates the percentage of studies where only good outcome patients are considered. Finally, it calculates the good to poor outcome ratio for these studies and plots its distribution.

```python
# Filter rows with both 0 or NaN values in both columns (non-operated cohorts)
filtered_df = design_details_df[(design_details_df["Cohort Size - Good outcomes\n[Question ID: 57634][Field ID: 93761x122146]"].isin([0.0])) &
                                (design_details_df["Cohort Size - Poor outcomes\n[Question ID: 57634][Field ID: 93762x122147]"].isin([0.0])) |
                                (design_details_df["Cohort Size - Good outcomes\n[Question ID: 57634][Field ID: 93761x122146]"].isna()) &
                                (design_details_df["Cohort Size - Poor outcomes\n[Question ID: 57634][Field ID: 93762x122147]"].isna())]
# Calculate the percentage of studies
percentage_non_operated = (len(filtered_df) / len(design_details_df)) * 100
print(f"Percentage of studies using non-operated patients undergoing iEEG monitoring: {percentage_non_operated:.2f} %")

# Filter only rows with numeric values
filtered_df_integers = design_details_df[(pd.to_numeric(design_details_df["Cohort Size - Good outcomes\n[Question ID: 57634][Field ID: 93761x122146]"], errors='coerce').notnull()) &
                                         (pd.to_numeric(design_details_df["Cohort Size - Poor outcomes\n[Question ID: 57634][Field ID: 93762x122147]"], errors='coerce').notnull())]

# Calculate the percentage of studies with good outcome patients only
filtered_df_good_outcome = filtered_df_integers[(filtered_df_integers["Cohort Size - Good outcomes\n[Question ID: 57634][Field ID: 93761x122146]"] > 0.0) &
                                                (filtered_df_integers["Cohort Size - Poor outcomes\n[Question ID: 57634][Field ID: 93762x122147]"] == 0.0)]
percentage_good_outcome_only = (len(filtered_df_good_outcome) / len(design_details_df)) * 100
print(f"Percentage of studies considering only good outcome patients: {percentage_good_outcome_only:.2f} %")

# Calculate the good to poor outcome ratio for the studies meeting the condition
both_outcomes = filtered_df_integers.drop(filtered_df_good_outcome.index)

outcome_ratios = both_outcomes["Cohort Size - Good outcomes\n[Question ID: 57634][Field ID: 93761x122146]"] / both_outcomes["Cohort Size - Poor outcomes\n[Question ID: 57634][Field ID: 93762x122147]"]

# Plot the distribution of good to poor outcome ratio
plt.figure(figsize=(8, 4))
outcome_ratios.hist(bins=50, color='grey', edgecolor='white')
plt.xlabel("Good to Poor Outcome Ratio")
plt.ylabel("Frequency")
plt.grid(True)

# Calculate the median, Q1, and Q3
median_value = outcome_ratios.median()
Q1 = outcome_ratios.quantile(0.25)
Q3 = outcome_ratios.quantile(0.75)

# Plot the vertical lines for median, Q1, and Q3
plt.axvline(median_value, color='red', linestyle='-', linewidth=2, label=f'Median: {median_value:.2f}')
plt.axvline(Q1, color='red', linestyle='--', linewidth=2, label=f'Q1: {Q1:.2f}')
plt.axvline(Q3, color='red', linestyle='--', linewidth=2, label=f'Q3: {Q3:.2f}')

plt.legend()
plt.show()
```

The code below reads the possible options directly from the "Cohort Selection Criteria" column and then print a table and plot a pie chart to visualize these percentages. Adjust the DataFrame df to match your actual dataset.

```python
# Initialize an empty dictionary to store counts of each option
option_counts = {}

# Iterate through each row and update the counts
for row in design_details_df["Cohort Selection Criteria\n[Question ID: 57635][Field ID: 97180x126675]"]:
    if isinstance(row, str):  # Handle single strings
        # Split the string by ", " and check for specific options
        row_options = [option.strip() for option in row.split("\n")]
        
        # Handle cases where "Consecutive" is present
        if any(option == "Consecutive" for option in row_options):
            option_counts["Consecutive"] = option_counts.get("Consecutive", 0) + 1
        # Handle cases where "Unclear" is present along with other options
        elif any(option == "Unclear" for option in row_options):
            option_counts["Unclear"] = option_counts.get("Unclear", 0) + 1
        # Handle cases where "Non-consecutive" and "Non-operated" are present together
        elif all(option in row_options for option in ["Non-consecutive", "Non-operated"]):
            option_counts["Non-consecutive"] = option_counts.get("Non-consecutive", 0) + 1
        elif len(row_options)==1 and row_options[0]=="Non-operated":
            option_counts["Unclear"] = option_counts.get("Unclear", 0) + 1 # one study was found to have "Non-operated" without saying if it's consecutive or not; this was checked afterwards and the selection criteria were unclear so we manually define this
        else:
            # Handle all other cases
            for option in row_options:
                option_counts[option] = option_counts.get(option, 0) + 1

# Convert counts to percentages
total_rows = len(design_details_df)
option_percentages = {option: (count / total_rows) * 100 for option, count in option_counts.items()}

# Create a DataFrame for visualization
percentage_design_details_df = pd.DataFrame(option_percentages.items(), columns=["Option", "Percentage"])

# Print the table
print("Percentage of each option in Cohort Selection Criteria:")
print(percentage_design_details_df)

# Plot the pie chart
colors = sns.color_palette('Set3', 10)  # Choose a palette and the number of distinct colors you need

plt.figure(figsize=(4, 5))
plt.pie(percentage_design_details_df["Percentage"], labels=percentage_design_details_df["Option"], colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Cohort Selection Criteria")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
# Function to count the number of epilepsy types in a study
def count_epilepsy_types(row):
    if isinstance(row, str):
        return len(row.split("\n"))
    else:
        return 0

# Apply the function to each row to get the number of epilepsy types considered
epi_types_cnt = design_details_df["Epilepsy Types\n[Question ID: 57637][Field ID: 93700x122085]"].apply(count_epilepsy_types)

# Calculate the percentage of studies for each number of epilepsy types
percentage_by_num_types = epi_types_cnt.value_counts(normalize=True) * 100

print("Percentage of studies by number of epilepsy types:")
print(percentage_by_num_types)

# Create a DataFrame
df_epi_types_count = pd.DataFrame({"Num_Epilepsy_Types": list(percentage_by_num_types.keys()),
                   "Percentage": list(percentage_by_num_types)})
# Sort the DataFrame by the 'Number of Epilepsy Types' column
sorted_df_epi_types_count = df_epi_types_count.sort_values(by='Num_Epilepsy_Types')
# Plot pie chart
plt.figure(figsize=(5, 6))
plt.pie(sorted_df_epi_types_count["Percentage"], labels=sorted_df_epi_types_count["Num_Epilepsy_Types"].astype(str), colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Percentage of Studies by Number of Epilepsy Types")
plt.axis('equal')
plt.show()

# Filter out the row with a value of 0 in the "Number of Epilepsy Types" column
filtered_df = epi_types_cnt[epi_types_cnt != 0]

# Histogram
plt.figure(figsize=(3, 4))
filtered_df.hist(bins=7, color='grey', edgecolor='white')
plt.xlabel("Number of Epilepsy Types")
plt.ylabel("Frequency")
# Set xticks
plt.xticks(range(1, 8))
plt.grid(True)

# Calculate the median, Q1, and Q3
median_value = filtered_df.median()
Q1 = filtered_df.quantile(0.25)
Q3 = filtered_df.quantile(0.75)

# Plot the vertical lines for median, Q1, and Q3
plt.axvline(median_value, color='red', linestyle='-', linewidth=2, label=f'Median: {median_value:.2f}')
plt.axvline(Q1, color='red', linestyle='--', linewidth=2, label=f'Q1: {Q1:.2f}')
plt.axvline(Q3, color='red', linestyle='--', linewidth=2, label=f'Q3: {Q3:.2f}')

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

The "Outcome" column will indicate whether "outcome" is present in the performance score column (1 if present, 0 otherwise), and the "Keywords" column will indicate whether any of the keywords are present in the abstract (1 if present, 0 otherwise).

```python
# Load the abstrackr DataFrame
abstrackr_df = pd.read_excel(file_path + "abstrackr_consensus_090424.xlsx")

# List of keywords to search for in abstracts
keywords = ["surgical outcome", "seizure freedom"]

# Initialize an empty DataFrame to store the results
results_df = pd.DataFrame(columns=["pubmed id", "Outcome", "Keywords"])

# Initialize an empty list to store the rows
rows = []

# Iterate through each study in the design details DataFrame
for index, row in design_details_df.iterrows():
    # Check if the study has "outcome" in the performance score column
    if "outcome" in str(row["Performance Score - Metric\n[Question ID: 57631][Field ID: 93691x122076]"]).lower():
        outcome = 1
    else:
        outcome = 0
    
    # Get the PMID of the study
    study_pmid = row["PMID"]
    
    # Check if the PMID exists in the abstrackr DataFrame
    if study_pmid in abstrackr_df["pubmed id"].values:
        # Get the abstract corresponding to the PMID
        abstract = abstrackr_df.loc[abstrackr_df["pubmed id"] == study_pmid, "abstract"].values[0]
        
        # Check if any of the keywords are present in the abstract
        if any(keyword in abstract.lower() for keyword in keywords):
            keywords_present = 1
        else:
            keywords_present = 0
    else:
        # If PMID not found in abstrackr, set keywords_present to 0
        keywords_present = 0

    # Append the row to the list
    rows.append({"pubmed id": study_pmid, "Outcome": outcome, "Keywords": keywords_present})

# Create a DataFrame from the list of rows
results_df = pd.concat([results_df, pd.DataFrame(rows)])

# Assuming results_df is your DataFrame containing the data
outcome_0_with_keywords_1 = results_df[(results_df['Outcome'] == 0) & (results_df['Keywords'] == 1)]

# Count the number of rows
count_outcome_0_with_keywords_1 = len(outcome_0_with_keywords_1)

print("Number of Outcome 0 with Keywords 1:", count_outcome_0_with_keywords_1)
```

Extract the most frequent words from the abstracts of studies where the Outcome column is 1. These words will then be used as keywords for further analysis. Adjust the value of n to control the number of keywords extracted.

```python
from collections import Counter

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
# # Download NLTK resources if not already downloaded
# import nltk
# nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

# Filter rows where Outcome is 1
outcome_studies = results_df.loc[results_df["Outcome"] == 1]

# Initialize an empty list to store abstracts
abstracts = []

# Iterate through the rows and fetch abstracts for studies with Outcome 1
for index, row in outcome_studies.iterrows():
    pmid = row["pubmed id"]
    abstract_row = abstrackr_df.loc[abstrackr_df["pubmed id"] == pmid]
    if not abstract_row.empty:
        abstract = abstract_row.iloc[0]["abstract"]
        if isinstance(abstract, str):
            abstracts.append(abstract)

# Function to preprocess text and extract bigrams and trigrams
def extract_ngrams(texts, n=2):
    all_ngrams = []
    for text in texts:
        # Tokenize text
        tokens = word_tokenize(text.lower())
        # Filter out stop words and punctuation
        filtered_tokens = [token for token in tokens if token.isalnum() and token not in stop_words]
        # Extract ngrams
        ngrams_list = list(ngrams(filtered_tokens, n))
        all_ngrams.extend(ngrams_list)
    return all_ngrams

# Extract bigrams and trigrams from abstracts with Outcome 1
outcome_1_ngrams = extract_ngrams(abstracts, n=2) + extract_ngrams(abstracts, n=3)

# Count the frequency of each bigram and trigram
ngram_counts = Counter(outcome_1_ngrams)

# Get the most common bigrams and trigrams
HOW_MANY = 20
most_common_ngrams = ngram_counts.most_common(HOW_MANY)
print(most_common_ngrams)
```

```python
# Read the Excel file into a DataFrame
df_design = pd.read_excel(file_path+file_name, sheet_name="Design Details")
df_rob = pd.read_excel(file_path+file_name, sheet_name="Risk of Bias Assessment")

# Define categories and corresponding keywords
categories = {
    "LFP": ["high frequency oscillations", "high-gamma", "wavelets", "EI", "slow wave activity", "univariate periodogram-type power measure", "HFOs", "HFO", "FA", "ripple", "ripples", "spike", "spikes", "gamma oscillations", "gamma", "IED", "PAC", "RMS", "fast", "CFC", "spatial dispersion"],
    "Connectivity analysis": ["connectivity", "networks", "CM", "synchronizability", "PLV", "FC", "fragility", "information flow", "network characteristics", "granger causality", "SCI", "graph", "centrality", "eigenvector", "energy", "transfer function", "coherence"],
    "Model": ["model", "modeling", "modelling", "dynamical", "state", "cross-sectional", "VEP"],
    "ML": ["ML", "machine learning", "CNN", "LOOCV", "CV", "cross validation", "cross-validation", "LOO", "ANN", "conv", "neural network", "artificial", "classification", "classified", "logistic regression", "GLM", "RF", "random forest", "SVM", "deep learning"],
    "ESI": ["ESI", "source localization", "source imaging"],
    "Stimulation": ["CCEP", "cortico-cortical", "SPES", "stimulation"]
}

# Initialize counters for each category
category_counts = {category: [] for category in categories}

# Loop through each row of the DataFrame
for index, row in df_design.iterrows():
    approach = str(row["Model / Analytical approach\n[Question ID: 57624][Field ID: 93677x122062]"]).lower()  # Convert to lowercase for case-insensitive matching
    citation_name = str(row["Citation Name"]).lower()  # Convert to lowercase for case-insensitive matching
    index_test = df_rob.at[index, "Review the index test and how it was conducted and interpreted. - Notes/Comments:\n[Question ID: 57907][Field ID: 94150x122821]"]

    # Check keywords for each category
    for category, keywords in categories.items():
        # Check if any keyword from the category is present in either approach or citation name
        # We add a space " " after the keyword to prevent finding keywords within other words
        if not type(index_test)==str:
            if any(keyword.lower()+" " in approach or keyword.lower()+" " in citation_name for keyword in keywords):
                # If found, increment count for the category
                category_counts[category].append((row["Citation Name"], row["Model / Analytical approach\n[Question ID: 57624][Field ID: 93677x122062]"]))
        else:
            if any(keyword.lower()+" " in approach or " "+keyword.lower() in approach or keyword.lower()+" " in citation_name or " "+keyword.lower() in citation_name or keyword.lower()+" " in index_test or " "+keyword.lower() in index_test for keyword in keywords):
                # If found, increment count for the category
                category_counts[category].append((row["Citation Name"], row["Model / Analytical approach\n[Question ID: 57624][Field ID: 93677x122062]"]))

# Print the counts for each category
for category, count in category_counts.items():
    print(f"{category}: {len(count)} studies")

# Visualize the results
plt.bar(category_counts.keys(), list(len(pub) for pub in category_counts.values()), color="grey")
plt.ylabel("Number of Studies", fontsize=15)
plt.title("Summary of Study Approaches", fontsize=15)
plt.xticks(rotation=45, ha="right", fontsize=15)
plt.tight_layout()
plt.show()
```
