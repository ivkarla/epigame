```python
def get_connectivity(matrix, row, col):
  # Get the connectivity measure for a node pair with indices row and col

    if row < 0 or row >= len(matrix) or col < 0 or col >= len(matrix[0]):
        return None  # Indices are out of bounds
    return matrix[row][col]
```

```python
def create_index_dict(result):
  # Create a dictionary to index tuples by their first value (node pair indices)
  index_dict = {}
  for item in result:
      first_value = item[0]
      if first_value in index_dict:
          print("Error: Pair duplicate.")
      else:
          index_dict[first_value] = item
  return index_dict

def get_top_epochs(result_index_dict, node1, node2, all_epochs, sigmas=1):
  # Get epochs that were used as training data for folds with CVS above threshold
  train_epochs = []
  cvs = result_index_dict[(node1, node2)][2]
  val_epochs = result_index_dict[(node1, node2)][3]

  # Calculate mean and standard deviation of the cvs list
  mean_cvs = np.mean(cvs)
  std_deviation_cvs = np.std(cvs)
  threshold = mean_cvs + std_deviation_cvs*sigmas

  # Find indices of cvs values above mean + standard deviation
  indices_above_threshold = [i for i, cv in enumerate(cvs) if cv > threshold]
  # Get corresponding val_epochs values using the indices
  val_epochs_above_threshold = [val_epochs[i] for i in indices_above_threshold]

  for val_eps in val_epochs_above_threshold:
    # Remove values that are in val_epochs_above_threshold list
    train_epochs.append([epoch for epoch in all_epochs if epoch not in val_eps])

  # Create an empty set to hold the union of all elements
  union_set = set()
  # Iterate through each sublist and update the union_set
  for sublist in train_epochs: union_set.update(sublist)
  # Convert the union_set back to a list
  top_epochs = list(union_set)

  return top_epochs
```

```python
from scipy.stats import norm

def variance_ratio_test(time_series, num_splits=2):
    n = len(time_series)
    k = n // num_splits

    split_series = np.array_split(time_series, num_splits)

    r_s = np.sum([(split_series[i][-1] - split_series[i][0]) ** 2 for i in range(num_splits)])
    s_s = np.sum([np.var(split_series[i]) for i in range(num_splits)])

    z = (r_s / (s_s * k)) - 1
    p_value = 1 - norm.cdf(z)

    return z, p_value
```

```python
from scipy import stats

def stat_test(sample1, sample2):
    # Perform a normality test (e.g., Shapiro-Wilk)
    normality_test_statistic1, normality_p_value1 = stats.shapiro(sample1)
    normality_test_statistic2, normality_p_value2 = stats.shapiro(sample2)

    # Define significance level
    alpha = 0.05

    # Check if data is Gaussian
    if normality_p_value1 > alpha and normality_p_value2 > alpha:
        # Data is approximately Gaussian, apply Z-test (assuming known population std. dev.)
        
        # Calculate the mean and standard deviation of the samples
        mean1 = np.mean(sample1)
        mean2 = np.mean(sample2)
        std1 = np.std(sample1)
        std2 = np.std(sample2)
        
        # Calculate the Z-score
        z_score = (mean1 - mean2) / ((std1**2 / len(sample1)) + (std2**2 / len(sample2)))**0.5
        
        # Calculate the two-tailed p-value
        p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
        
    else:
        # Data is not Gaussian, apply Mann-Whitney U test
        U_statistic, p_value = stats.mannwhitneyu(sample1, sample2)
    return p_value
```
