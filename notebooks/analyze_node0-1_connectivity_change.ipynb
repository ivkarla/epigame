```python
# Sample groups of size equal to 10% of total nodes

r = 0.1
group_size = int(len(NODES[str(random_sub)])*r) if int(len(NODES[str(random_sub)])*r) < len(RESECTION[random_sub]) else len(RESECTION[random_sub])

print(group_size, int(len(NODES[str(random_sub)])*r) < len(RESECTION[random_sub]))

resection_labels = list(np.random.choice(RESECTION[random_sub], size=group_size, replace=False))
non_resection_labels = list(np.random.choice(np.setdiff1d(NODES[str(random_sub)], resection_labels), size=group_size, replace=False))

# Get the first label in the '<->' separated string
first_labels = df_sub['Labels'].str.split('<->').str[0]

# Filter rows where the first label is in resection_labels
resection_rows = df_sub[first_labels.isin(resection_labels)]

# Filter rows where the second label is in either resection_labels or non_resection_labels
second_labels = df_sub['Labels'].str.split('<->').str[1]

relevant_rows = resection_rows[second_labels.isin(resection_labels + non_resection_labels)]

# Print the relevant rows and their Average_CVS values

# for index, row in relevant_rows.iterrows():
#     print(f"Labels: {row['Labels']}, Average_CVS: {row['Average_CVS']}")

# Filter data for different conditions

condition_10 = (relevant_rows['0_Node_Type'] == 1) & (relevant_rows['1_Node_Type'] == 0)
condition_11 = (relevant_rows['0_Node_Type'] == 1) & (relevant_rows['1_Node_Type'] == 1)

plot_condition_dist(relevant_rows, condition_10, condition_11)
```

```python
# Calculate average CVS values for conditions 10 and 11
avg_cvs_condition_10 = relevant_rows[condition_10]['Average_CVS']
avg_cvs_condition_11 = relevant_rows[condition_11]['Average_CVS']

# Calculate mean and standard deviation
mean_condition_10 = avg_cvs_condition_10.mean()
std_condition_10 = avg_cvs_condition_10.std()

mean_condition_11 = avg_cvs_condition_11.mean()
std_condition_11 = avg_cvs_condition_11.std()

# Calculate thresholds based on sigma
thresholds = [1, 2, 3, 4]

thresholded_avg_cvs_condition_10 = {threshold: [avg for avg in avg_cvs_condition_10 if avg > mean_condition_10 + std_condition_10 * threshold] for threshold in thresholds}
thresholded_avg_cvs_condition_11 = {threshold: [avg for avg in avg_cvs_condition_11 if avg > mean_condition_11 + std_condition_11 * threshold] for threshold in thresholds}

print("Thresholded Average CVS values for condition_10:")
print(thresholded_avg_cvs_condition_10)

print("Thresholded Average CVS values for condition_11:")
print(thresholded_avg_cvs_condition_11)
```

```python
# Create paired boxplots for each threshold
plt.figure(figsize=(12, 6))

for idx, threshold in enumerate(thresholds):
    plt.subplot(1, len(thresholds), idx + 1)
    data_to_plot = [thresholded_avg_cvs_condition_10[threshold], thresholded_avg_cvs_condition_11[threshold]]
    plt.boxplot(data_to_plot, labels=['1-0', '1-1'])
    plt.title(f"Average CVS > mean+{threshold}σ")
    plt.ylabel('Average CVS')

plt.tight_layout()
plt.show()
```

```python
import itertools

for sub in good_outcome_subjects:

    print(f"Subject: {sub}")
    df_sub = df[df['Subject'] == sub]

    r = 0.1
    group_size = int(len(NODES[str(sub)])*r) if int(len(NODES[str(sub)])*r) < len(RESECTION[sub]) else len(RESECTION[sub])
    print(group_size, int(len(NODES[str(random_sub)])*r) < len(RESECTION[random_sub]))

    avg_cvs_condition_10, avg_cvs_condition_11, avg_cvs_condition_01 = [],[],[]

    for i in range(10):

        resection_labels = list(np.random.choice(RESECTION[sub], size=group_size, replace=False))
        non_resection_labels = list(np.random.choice(np.setdiff1d(NODES[str(sub)], resection_labels), size=group_size, replace=False))

        for en in [RESECTION[sub][0]]:

            # Split the Labels column into two nodes using '<->'
            split_labels = df_sub['Labels'].str.split('<->', expand=True)

            # Check if one node is the en and the other is in resection_labels or non_resection_labels
            relevant_rows = df_sub[(split_labels[0] == en) & split_labels[1].isin(resection_labels + non_resection_labels) |
                              (split_labels[1] == en) & split_labels[0].isin(resection_labels + non_resection_labels)]

            condition_10 = (relevant_rows['0_Node_Type'] == 1) & (relevant_rows['1_Node_Type'] == 0)
            condition_01 = (relevant_rows['0_Node_Type'] == 0) & (relevant_rows['1_Node_Type'] == 1)
            condition_11 = (relevant_rows['0_Node_Type'] == 1) & (relevant_rows['1_Node_Type'] == 1)

            # Calculate Max CVS values for conditions 10 and 11
            avg_cvs_condition_10 += list(relevant_rows[condition_10]['CVS'].apply(lambda x: list(map(float, x.strip('[]').split()))))
            avg_cvs_condition_10 += list(relevant_rows[condition_01]['CVS'].apply(lambda x: list(map(float, x.strip('[]').split()))))
            avg_cvs_condition_11 += list(relevant_rows[condition_11]['CVS'].apply(lambda x: list(map(float, x.strip('[]').split()))))

            print(len(avg_cvs_condition_10), len(avg_cvs_condition_11))

    # Calculate mean and standard deviation
    mean_condition_10 = np.mean(avg_cvs_condition_10)
    std_condition_10 = np.std(avg_cvs_condition_10)

    mean_condition_11 = np.mean(avg_cvs_condition_11)
    std_condition_11 = np.std(avg_cvs_condition_11)

    # Calculate thresholds based on sigma
    thresholds = [1, 2, 3, 4]

    # Flatten the avg lists using itertools.chain.from_iterable
    flattened_avg_cvs_condition_10 = list(itertools.chain.from_iterable(avg_cvs_condition_10))
    flattened_avg_cvs_condition_11 = list(itertools.chain.from_iterable(avg_cvs_condition_11))

    # Calculate thresholded average CVS values for condition_10
    thresholded_avg_cvs_condition_10 = {
        threshold: [val for val in flattened_avg_cvs_condition_10 if val > mean_condition_10 + std_condition_10 * threshold]
        for threshold in thresholds
    }

    # Calculate thresholded average CVS values for condition_11
    thresholded_avg_cvs_condition_11 = {
        threshold: [val for val in flattened_avg_cvs_condition_11 if val > mean_condition_11 + std_condition_11 * threshold]
        for threshold in thresholds
    }
    # Create paired boxplots for each threshold
    fig, axs = plt.subplots(1, len(thresholds) + 1, figsize=(15, 6))

    for idx, threshold in enumerate(thresholds):
        axs[idx].boxplot([thresholded_avg_cvs_condition_10[threshold], thresholded_avg_cvs_condition_11[threshold]], labels=['1-0', '1-1'])
        axs[idx].set_title(f"Max CVS > mean+{threshold}σ")
        axs[idx].set_ylabel('Max CVS')

    # Add a subplot for distributions
    axs[-1].hist(relevant_rows.loc[condition_10, 'Max_CVS'], bins=10, alpha=0.5, label='0_Node_Type=1, 1_Node_Type=0')
    axs[-1].hist(relevant_rows.loc[condition_11, 'Max_CVS'], bins=10, alpha=0.5, label='0_Node_Type=1, 1_Node_Type=1')
    axs[-1].set_xlabel('Max CVS')
    axs[-1].set_ylabel('Frequency')
    axs[-1].set_title('Histogram')
    axs[-1].legend(loc='upper left', bbox_to_anchor=(1, 1))

    plt.tight_layout()
    plt.show()
```
